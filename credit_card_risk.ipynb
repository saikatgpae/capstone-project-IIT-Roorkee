As data set is highli imbalanced, we will create a balanced sample data set from the main data set to train the model.
Building train data set.
284315 Legit or normal transaction.
492 fraud transaction.
So we will randomly choose 492 normal transaction to match with 492 fraud transaction to prepare a balanced sample data set for a better prediction.

```python
#creating 492 random legit samples
legit_sample = legit.sample(n=492)
```

```python
# creating new data frame by concating legit_sample with fraud class
new_df = pd.concat([legit_sample, fraud], axis=0)
```

```python
#checking the head of new data frame
new_df.head()
```

```python
#checking the tail of the new data frame
new_df.tail()
```

```python
#counting of both legit and fraud transaction
new_df['Class'].value_counts()
```

```python
#comparing the values of both legit and fraud transaction
new_df.groupby('Class').mean()
```

Spliting the data into Features & Targets

```python
#dropping the predicting features, The class column
X = new_df.drop(columns='Class', axis=1)
Y = new_df['Class']
```

```python
print(X)
```

```python
print(Y)
```

<b>Splitting the Data into Training and Testing Data</b>

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
#checking the spliting
print(X.shape, X_train.shape, X_test.shape)
```

<b>Modele Traing with Logistic Regression</b>

```python
#creating hte model
model = LogisticRegression()
```

```python
# training the model with training data
model.fit(X_train, Y_train)
```

<b>Model Evalution</b>

```python
# finding the accuracy score of training data
X_train_predict = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predict, Y_train)
#printing the accuracy of training data
print('Accuracy of Training data = ',training_data_accuracy)
```

Prediction is good of Traing data, more than 75%

```python
# finding the accuracy score of test data
X_test_predict = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_predict, Y_test)
#printing the accuracy of test data
print('Accuracy of Test data = ',testing_data_accuracy)
```

Prediction is good of Testing data, more than 75%
